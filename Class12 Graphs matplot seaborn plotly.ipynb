```python
 #What are cufflinks in machine learning?
#Cufflinks connect Plotly with pandas to create graphs and charts of Dataframes directly. 
#Its a Python library which is used to design graphs, especially interactive graphs. 
#It can plot various graphs and charts like histogram, bar plot, boxplot, spread plot and many more.
import cufflinks as cf
cf.go_offline()
```

```python
cf.set_config_file(theme="polar")
df.iplot
```

```python
df.iplot(x = "sepal_length", y = 'sepal_width', kind = "scatter", mode = 'markers', xTitle = 'sepal_length', yTitle = 'sepal_width')
```
