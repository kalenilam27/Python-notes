```python
data3 = { "emp_id":[101, 102, 103, 104 ,105] ,
         "salary": [10000,2000,3000,4000,5000],
         "phone_no" :[12345 ,678,9087,6754,3214]}
```

```python
data4= {"emp_id":[101, 102, 103, 104 ,106],
       "course" : ["eng","hindi","maths","social","evs"],
       "course_ID":[1,2,3,4,5]}
```

```python
df4 = pd.DataFrame(data3)
```

```python
df5 = pd.DataFrame(data4)
```

```python
pd.merge(df4,df5)
```

```python
pd.merge(df4,df5, how ="right")
```

```python
pd.merge(df4,df5, how ="left")
```

```python
pd.merge(df4,df5, how ="right" , on = "emp_id")
```

```python
pd.merge(df4,df5, how ="left" , on = "emp_id")
```

```python
data5 = {"emp_id1":[101,102,103,104,105],
        "salary": [10000,2000,3000,4000,5000],
         "phone_no" :[12345 ,678,9087,6754,3214]}

data6 = {"emp_id2":[101,102,103,106,107],
        "course" : ["eng","hindi","maths","social","evs"],
       "course_ID":[1,2,3,4,5]}
```

```python
df6 = pd.DataFrame(data5)
df7 = pd.DataFrame(data6)
```

```python
pd.merge(df6,df7, left_on = "emp_id1", right_on = "emp_id2")
```
